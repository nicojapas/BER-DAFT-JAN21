# Activity 3

```python
for i in [10, 100, 1000, 10000]:
    plt.figure()
    n_of_experiments=1000
    x = np.random.binomial(n=i, p=0.5, size=n_of_experiments)
    
    # what happens with the mean?
    mean_x = np.mean(x)
    
    sns.displot(x)
    
    plt.alxvine(mean_x, c='red')
    plt.title(f'Mean: {round(mean_x,2)} heads out of {n_of_experiments}\n experiments with {i} tosses each')
    plt.show()
```

Law of large numbers!
The law of large numbers, in probability and statistics, states that as a sample size grows, its mean gets closer to the average of the whole population.

# Activity 4

Let's visualize the **Central limit theorem** and sample from a weird distribution

```python
import matplotlib.pyplot as plt
import random
```

```python
# create a weird bimodal distribution
bimodal1 = np.random.normal(loc=90, scale=15, size=50000)
bimodal2 = np.random.normal(loc=20, scale=15, size=50000)

bimodal_test = np.concatenate((bimodal1, bimodal2))

# plot it
plt.hist(bimodal_test, bins=200);
```

```python
# Let's sample that 50 times (independently!
# identically distributed, IID) and calculate the mean

def plt_means(data, n_times):
    means = []
    for i in tqdm(range(n_times)):
        sample_pi = random.sample(set(data), 50)
        means.append(np.mean(sample_pi))
    plt.hist(means, bins=20)
    
plt_means(bimodal_test, 50)
```

```python
plt_means(bimodal_test, 500)
```

```python
plt_means(bimodal_test, 5000)
```

# Central Limit theorem (one definition out of many)

*If we know the population mean and standard deviation, we know the following will be true:*

> The **distribution of means** across repeated samples will be **normal** with a mean equal to the **population mean** (and a standard deviation equal to the population standard deviation divided by the square root of n).

# Why is Central limit theorem so cool?

* There are various definitions, but all lead to the same statement. In case you google it
* Lindeberg Levy CLT

* Creates a connection between **our sample** and the **real world outside**!
* The more and the larger the samples are, the better we can infer from sample to population!
* that means: **We can infer from our samples to the ground truth (population!)**
* and that means: **We don't need to care about the distribution of the population!**
* Imagine you make 1 sample, get a mean of $\bar{x}$ = 50. Then you can answer a question of "How likely is it, that I have measured this exact mean", because we know the underlying distribution of those means!
* or in other words: We can test the Hypothesis, that our sample represents a population distinct from the known population!

* argument: But we never reach an infinite amount of samples!
    * Yes! But: We reach a normal distribution of really small sample size!

### Again, sampling from an exponential distribution!

```python
sns.displot(np.random.exponential(size=1000))
```

```python
# a function wich samples the distribution rep times, gathers the means and returns them as a list
def sample_from_distr(sample_size, rep):
    return [np.random.exponential(size=sample_size).mean() for r in range(rep)]
```

```python
# plot this list
sns.displot(sample_from_distr(1000,1000))
```

It's a normal distribution!
